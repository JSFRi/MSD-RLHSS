## Maximal temperature

```python
all_tier
```

```python
def hot_cold(tier,file):
    tier.loc[tier['No.']==file['No.'].item(),'req_time']+=1
    file=tier.loc[tier['No.']==file['No.'].item()]
    ## weight/request_time as probability ratio
    ratio=1-1.8/(1+np.exp(5000*7.33/100000)*np.exp(-(7.33/100000)*float(file['weight']/file['req_time'])))
    
    ##if file is cold, change it to hot according to the weight/req_time ratio
    if float(file['temp'])<0.5:
        if random.random()<ratio:
            temp=1-0.5/np.exp(0.1*file['req_time'])
            file['temp']=temp
            tier.loc[tier['No.']==file['No.'].item()]=file
        else:
            pass
     ##if the file is hot, increase temperatue by requests time
    else:
        temp=1-0.5/np.exp(0.1*file['req_time'])
        file['temp']=temp
        tier.loc[tier['No.']==file['No.'].item()]=file

## Naturally decreasement of temp
def temp_decrease(tier1,tier2,tier3,Request,timestep):
    
    req_No=list(Request.loc[Request['request']==1,'No.'])
    
    for tier in [tier1,tier2,tier3]:
        for i in list(tier['No.']):
            if i in req_No:
                tier.loc[tier['No.']==i,'time_untouched']=0
            else:
                ##increase untouched time by 1
                tier.loc[tier['No.']==i,'time_untouched']+=1
                ## decrease temp by 0.1 after $timestep times untorched
                if int(tier.loc[tier['No.']==i,'time_untouched'])!=0 and int(tier.loc[tier['No.']==i,'time_untouched'])%timestep==0:
                    tier.loc[tier['No.']==i,'temp']-=0.1
                if float(tier.loc[tier['No.']==i,'temp'])<0.1:
                    tier.loc[tier['No.']==i,'temp']=0.1
```

```python
ti_tmp=[]
tier3=all_tier[0:0]
tier2=all_tier[0:0]
tier1=all_tier
for turn in range(600):
    req=pd.read_csv('./Requests_angel030/req_%d.csv'%turn)
    req_1=req.loc[req['request']==1]
    ti=0
    for no in req_1['No.']:
        if no in list(tier3['No.']):
            ti+=tier3.loc[tier3['No.']==no,'weight'].item()/100000
            file=tier3.loc[tier3['No.']==no]
            hot_cold(tier3,file)
        elif no in list(tier2['No.']):
            ti+=tier2.loc[tier2['No.']==no,'weight'].item()/50000
            file=tier2.loc[tier2['No.']==no]
            hot_cold(tier2,file)
        elif no in list(tier1['No.']):
            ti+=tier1.loc[tier1['No.']==no,'weight'].item()/10000
            file=tier1.loc[tier1['No.']==no]
            hot_cold(tier1,file)
        else:
            print('Error: no such file in tier3/2/1')
    temp_decrease(tier1,tier2,tier3,req,timestep=3)
    ti_tmp.append(ti)
    tier=pd.concat([tier3,tier2,tier1],ignore_index=True).sort_values(by='temp',ascending=False)
    i3=0
    while tier[0:i3]['weight'].sum() < 1000000:
        i3+=1
    tier3=tier[0:i3-1]
    #print(tier3['weight'].sum())
    i2=i3
    while tier[i3-1:i2]['weight'].sum() < 5000000:
        i2+=1
    tier2=tier[i3-1:i2-1]
    #print(tier2['weight'].sum())
    tier1=tier[i2-1:]
```

```python
plt.plot(ti_tmp)
```

```python
ti_tmp1=[]
tier3=all_tier[0:0]
tier2=all_tier[0:0]
tier1=all_tier
for turn in range(600):
    req=pd.read_csv('./Requests_angel030/req_%d.csv'%turn)
    req_1=req.loc[req['request']==1]
    ti=0
    print('Turn',turn)
    t=time.time()
    for no in req_1['No.']:
        if no in list(tier3['No.']):
            ti+=tier3.loc[tier3['No.']==no,'weight'].item()/100000
            file=tier3.loc[tier3['No.']==no]
            hot_cold(tier3,file)
        elif no in list(tier2['No.']):
            ti+=tier2.loc[tier2['No.']==no,'weight'].item()/50000
            file=tier2.loc[tier2['No.']==no]
            hot_cold(tier2,file)
            ## determine whether the file should be upgrade to tier3
            temp_not=tier3['temp'].sum()/len(tier3)
            tier3_new=pd.concat([tier3,file],ignore_index=True)
            temp_up=tier3_new['temp'].sum()/len(tier3_new)
            if temp_up > temp_not:
                tier3=tier3_new
            while tier3['weight'].sum() > 1000000:
                tier3=tier3.sort_values(by='temp',ascending=True)[1:]
        elif no in list(tier1['No.']):
            ti+=tier1.loc[tier1['No.']==no,'weight'].item()/10000
            file=tier1.loc[tier1['No.']==no]
            hot_cold(tier1,file)
            ## determine whether the file should be upgrade to tier2
            temp_not=tier2['temp'].sum()/len(tier2)
            tier2_new=pd.concat([tier2,file],ignore_index=True)
            temp_up=tier2_new['temp'].sum()/len(tier2_new)
            if temp_up > temp_not:
                tier2=tier2_new
            while tier2['weight'].sum() > 5000000:
                tier2=tier2.sort_values(by='temp',ascending=True)[1:]
        else:
            print('Error: no such file in tier3/2/1')
    temp_decrease(tier1,tier2,tier3,req,timestep=3)
    ti_tmp1.append(ti)
    print(time.time()-t,'seconds')
```
