## LRU

```python
all_tier=pd.read_csv('1000_Genomes_phase_3_release.tsv',sep='\t')
all_tier.loc[all_tier['Population code']=='IBS,MSL','Population code']='IBS'
all_tier['No.']=all_tier['Sample name']
all_tier['temp']=0.5
all_tier['weight']=1000
all_tier['req_time']=0
all_tier['time_untouched']=0
all_tier
```

```python
ti_lru=[]
tier3=all_tier[0:100]
tier2=all_tier[100:600]
tier1=all_tier[600:]
for turn in range(650):
    req=pd.read_csv('./Requests_population1vs25/req_%d.csv'%turn)
    req_1=req.loc[req['request']==1]
    ti=0
    for no in req_1['No.']:
        if no in list(tier3['No.']):
            ti+=20
        elif no in list(tier2['No.']):
            ti+=40
        elif no in list(tier1['No.']):
            ti+=100
        else:
            print('Error: no such file in tier3/2/1')
    ti_lru.append(ti)
    tier=pd.concat([all_tier.loc[all_tier['No.'].isin(req_1['No.'])].sample(frac=1) 
                    ,all_tier.loc[~all_tier['No.'].isin(req_1['No.'])].sample(frac=1)],ignore_index=True)
    tier3=tier[0:100]
    tier2=tier[100:600]
    tier1=tier[600:]
```

## K-means

```python
from sklearn.cluster import KMeans
```

```python
ti_kmeans=[]
all_tier_f=all_tier
all_tier_f['freq']=0
pop_dict=dict(zip(all_tier_f['Population code'].unique(),range(len(all_tier_f['Population code'].unique()))))
all_tier_f['code']=all_tier_f['Population code'].map(pop_dict)
tier3=all_tier_f[0:100]
tier2=all_tier_f[100:600]
tier1=all_tier_f[600:]
for turn in range(650):
    print('Turn',turn)
    req=pd.read_csv('./Requests_population1vs25/req_%d.csv'%turn)
    req_1=req.loc[req['request']==1]
    all_tier_f.loc[all_tier_f['No.'].isin(req_1['No.']),'freq']+=1
    ti=0
    for no in req_1['No.']:
        if no in list(tier3['No.']):
            ti+=20
        elif no in list(tier2['No.']):
            ti+=40
        elif no in list(tier1['No.']):
            ti+=100
        else:
            print('Error: no such file in tier3/2/1')
    ti_kmeans.append(ti)
    all_tier_f['label']=KMeans(n_clusters=100, random_state=0, n_init=10).fit(all_tier_f[['freq','code']]).labels_    
    tier=all_tier_f.sort_values(by=['freq','label'],ascending=False)
    tier3=tier[0:100]
    tier2=tier[100:600]
    tier1=tier[600:]
```

## Minimal Entropy

```python
entropy([30,20,10,2])
```

```python
entropy([31,20,10,2])
```

```python
ti_ent=[]
tier3=all_tier[0:100]
tier2=all_tier[100:600]
tier1=all_tier[600:]
for turn in range(650):
    req=pd.read_csv('./Requests_population1vs25/req_%d.csv'%turn)
    req_1=req.loc[req['request']==1]
    ti=0
    for no in req_1['No.']:
        if no in list(tier3['No.']):
            ti+=20
        elif no in list(tier2['No.']):
            ti+=40
            ## check if updating this file can reduce the entropy of tier3
            ent_not=entropy(list(tier3.groupby(by='Population code').count()['Sample name']))
            tier3_new=pd.concat([tier3,tier2.loc[tier2['No.']==no]],ignore_index=True)
            ent_up=entropy(list(tier3_new.groupby(by='Population code').count()['Sample name']))
            ## if true, then upgrade the file into tier3
            if ent_up < ent_not:
                tier3=pd.concat([tier3,tier2.loc[tier2['No.']==no]],ignore_index=True)
                ## if tier3 is full, downgrade the file with the lowest count pop code 
                if tier3['weight'].sum() > 100000:
                    ## lowest count pop code
                    idx=tier3.groupby(by='Population code').count()['Sample name'].sample(frac=1).idxmin()
                    ## downgrade file from tier3 to tier2
                    file=tier3.loc[tier3['Population code']==idx].sample(frac=1)[0:1]
                    tier3=tier3.drop(file.index).reset_index(drop=True)
                    tier2=pd.concat([tier2,file],ignore_index=True)
        elif no in list(tier1['No.']):
            ti+=100
            ## check if updating this file can reduce the entropy of tier2
            ent_not=entropy(list(tier2.groupby(by='Population code').count()['Sample name']))
            tier2_new=pd.concat([tier2,tier1.loc[tier1['No.']==no]],ignore_index=True)
            ent_up=entropy(list(tier2_new.groupby(by='Population code').count()['Sample name']))
            ## if true, then upgrade the file into tier3
            if ent_up < ent_not:
                tier2=pd.concat([tier2,tier1.loc[tier1['No.']==no]],ignore_index=True)
                ## if tier2 is full, downgrade the file with the lowest count pop code 
                if tier2['weight'].sum() > 500000:
                    ## lowest count pop code
                    idx=tier2.groupby(by='Population code').count()['Sample name'].sample(frac=1).idxmin()
                    ## downgrade file from tier2 to tier1
                    file=tier2.loc[tier2['Population code']==idx].sample(frac=1)[0:1]
                    tier2=tier2.drop(file.index).reset_index(drop=True)
                    tier1=pd.concat([tier1,file],ignore_index=True)
        else:
            print('Error: no such file in tier3/2/1')
    ti_ent.append(ti)
```
